```python
from pycaret.datasets import get_data
data = get_data('forest')
```

# Create pipeline

```python
from pycaret.regression import *
s = setup(data, target = 'DC', session_id = 123)
```

# Using API

```python
type(exp)
```

# Compare Models

```python
best = compare_models()
```
